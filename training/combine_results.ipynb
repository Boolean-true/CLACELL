# Combine results

```python
import pandas as pd
import os

folder = "results/autoencoder_lr"

all_runs_data = []

for i in range(10):
    file_path = os.path.join(folder, f"result_{i}.csv")
    df = pd.read_csv(file_path, header=[0, 1, 2, 3], index_col=0)
    all_runs_data.append(df)

df = pd.concat(all_runs_data, axis=0)

# Compute Statistics for Robustness Test results
means = df.mean()
stds = df.std()

combined = means.round(4).astype(str) + " +- " + stds.round(4).astype(str)

# Create Dataframe with Statistics
stats_df = pd.DataFrame(
    [means, stds, combined], 
    index=["mean", "std", "mean +- std"], 
    columns=df.columns
)

# Combine original Dataframe with Statistic Dataframe
final_df = pd.concat([df, stats_df], axis=0)

final_df.to_csv('results/autoencoder_lr/combined_results.csv', index=True)
```
